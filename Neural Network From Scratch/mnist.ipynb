- **The input layer has 784 values** Each image has 784 pixels. We have a matrix where each row has a length of 784. We transpose it so we have each column with 784 values, and the length of each row is the number of examples we have.
- The first hidden layer has 10 values
- The second hidden layer has 10 values too
- And the output layer has a number between 0 and 9

## The process to train this network has 3 steps

### First step: Forward propagation

Here yo take an image and run it through the network, you compute what the output is.

$A^{[0]} = x $ 

x is 784*m, where m equals the number of examples

$Z$ is the unactivated first layer. We get it by applying a weight and a bias.

$Z^{[1]} = W^{[1]} A^{[0]} + b^{[1]}$

$Z$ is $10*m$, $W$ is $10*784$ and $b$ is $10*1$

Then, we have to apply an activation function. But what happens is we don't have this step in our NN? 

Well, if the first layer is a linear combination of the input, and then between the first and the second layer all we have is another linear combination, finally as the computation of multiple linear combinations gives you just a new linear combination, the final output of our NN will be a linear combination of the input data. You'll never get an interesting function out of the NN (it will be just a really fancy linear regression)

To solve that we apply an activation function. After this the second layer will be able to add a new layer of complexity to the NN, so it is more powerfull. 

Here we will be using $ReLu(x)$ which is not linear

$ReLu(x) = \left\{\begin{array}{lr}
        x, & \text{x} > 0\\
        0, & \text{x} \leq 0
        \end{array}\right\}$

$A^{[1]} = ReLu(Z^{[1]})$

After this we compute the second layer.

$A^{[1]}$ is $10*m$

$Z^{[2]}$ is the unactivated first layer. We get it by applying a weight and a bias.

$Z^{[2]} = W^{[2]} A^{[1]} + b^{[2]}$

$Z^{[2]}$ is $10*m$, $W^{[2]}$ is $10*10$ and $b^{[2]}$ is $10*1$

The second activation function will be **Softmax** that gives us a probability vector based on the input vector. Bassicaly what the function does is to divide $e^{Z_{i}}$ by the sum of $e$ to every value:

$f(Z_{i}) = \frac {e^{Z_{i}}} {\displaystyle\sum_{j=1}^K e^{Z_{j}}}$

$A^{[2]} = f(Z^{[2]})$

And this function give us a vector with values between 0 and 1. The position with bigger value will be our predicted number. That finished first step.

## Second step: Back propagation

Back propagation goes the opposite way. 

We start with the output, and we're going to find out how much the prediction deviated by the actual label.

We're going to calculate how much each of the preious weights and biases contributed to that error. Then, we'll adjust those accordingly.

The error for our predictions:

$dZ^{[2]} = A^{[2]} - Y$ all in  $10*m$

with Y = 4 for example: $[0, 0, 0, 0, 1, 0, 0, 0, 0]$ (one hot codification)

Some fancy math to calculate the contribution of each W and b to the result:

$dW^{[2]} = {\frac{1}{m}}dZ^{[2]} A^{[1]T}$
