```python
import numpy as np

n = 5
#6 puntos equiespaciados en -1, 1
puntos = np.linspace(-1, 1, n+1)

def f(x):
    return 1/(1 + 25*x**2)

y = f(puntos)

A = np.vander(puntos, increasing=False)
A_inv = np.linalg.inv(A)
c = A_inv @ y
#coeficientes del polinomio
#p(x) = c[0] + c[1]*x + c[2]*x^2 + c[3]*x^3 + c[4]*x^4 + c[5]*x^5
c = c[::-1] #me había quedado dada vuelta la lista
def p(x):
    return sum(c[i]*x**i for i in range(len(c)))
#plot de p(x) and f(x)
import matplotlib.pyplot as plt
x = np.linspace(-1, 1, 100)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, p(x), label='p(x)')
plt.scatter(puntos, y, color='red')
plt.legend()
plt.show()
```

```python
#norma infinito
max(abs(f(x) - p(x)))
```
