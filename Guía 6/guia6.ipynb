# Punto 7

```python
import numpy as np

n = 5
#6 puntos equiespaciados en -1, 1
puntos = np.linspace(-1, 1, n+1)

def f(x):
    return 1/(1 + 25*x**2)

y = f(puntos)

A = np.vander(puntos, increasing=False)
A_inv = np.linalg.inv(A)
c = A_inv @ y
#coeficientes del polinomio
#p(x) = c[0] + c[1]*x + c[2]*x^2 + c[3]*x^3 + c[4]*x^4 + c[5]*x^5
c = c[::-1] #me había quedado dada vuelta la lista
def p(x):
    return sum(c[i]*x**i for i in range(len(c)))
#plot de p(x) and f(x)
import matplotlib.pyplot as plt
x = np.linspace(-1, 1, 100)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, p(x), label='p(x)')
plt.scatter(puntos, y, color='red')
plt.legend()
plt.show()
```

```python
#norma infinito
max(abs(f(x) - p(x)))
```

# Punto 8

```python
x = [1950, 1960, 1970, 1980, 1990, 2000]
#uso ln
x = np.log(x)
b = [17, 20.5, 23.9, 27.9, 32.6, 36.9]
b = np.log(b)

A = np.vander(x, increasing=True)

At = A.T

AtA = At @ A
print(AtA.shape, b.shape)
Ata_inv = np.linalg.inv(AtA)
Atb = At @ b

a = Ata_inv @ Atb

#polinomio
def p(t):
    return a[0] + a[1]*t + a[2]*t**2 + a[3]*t**3 + a[4]*t**4 + a[5]*t**5

#plot p
t = np.linspace(np.log(1950), np.log(2000), 100)
plt.plot(t, p(t), label='p(t)')
plt.scatter(x, b, color='red')
plt.legend()

#derivo p
def dp(t):
    return a[1] + 2*a[2]*t + 3*a[3]*t**2 + 4*a[4]*t**3 + 5*a[5]*t**4

print(dp(2000))
```

```python
x = [1950, 1960, 1970, 1980, 1990, 2000]
#uso ln
x = np.log(x)
b = [17, 20.5, 23.9, 27.9, 32.6, 36.9]
b = np.log(b)

A = np.vander(x, 2, increasing=True)

At = A.T

AtA = At @ A
Atb = At @ b

a = np.linalg.solve(AtA, Atb)

#polinomio
def q(t):
    return a[0] + a[1]*t

#plot p
t = np.linspace(np.log(1950), np.log(2000), 100)
plt.plot(t, q(t), label='p(t)')
plt.scatter(x, b, color='red')
plt.legend()

#derivo p
def dq(t):
    return a[1]

print(dq(2000))
```
